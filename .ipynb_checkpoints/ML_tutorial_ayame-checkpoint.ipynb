```python
# library
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

# データセットを導入
from sklearn.datasets import load_iris                # アヤメデータ
from sklearn.model_selection import train_test_split  # 学習とテスト用のデータを分割するやつ
from sklearn.svm import LinearSVC

flowers = load_iris()
```

```python
# pandasで整理
flowers_data = pd.DataFrame(flowers.data, columns=['がく片の長さ', 'がく片の幅', '花びらの長さ', '花びらの幅'])
flowers_target = pd.DataFrame(flowers.target, columns=['花の種類'])
flower_all = pd.concat([flowers_data, flowers_target], axis=1) # axis: 0-rows, 1-columns

# 出力(10行)
flower_all.head(10)
```

```python
# 出力(平均、最大値など)
flower_all.describe()
```

```python
# 学習用とテスト用にデータを分割
split_size = 0.33  #全体の33%をテスト用に使用
train_data, test_data, train_target, test_target = train_test_split(flowers_data, flowers_target, test_size=split_size)
```

```python
# 出力
train_data.head(5)
```

```python
train_target.head(5)
```

```python
train_data.describe()
```

```python
# train_target['花の種類']==0 と　train_data[(train_target == 0).values
# どちらを使用しても同じ
# 条件式が成り立つかがboolで返ってくる

# matplotlibで出力
plt.scatter(train_data[train_target['花の種類']==0]['がく片の長さ'], train_data[(train_target == 0).values]['花びらの幅'], label='setosa', cmap='rgb')
plt.scatter(train_data[train_target['花の種類']==1]['がく片の長さ'], train_data[(train_target == 1).values]['花びらの幅'], label='setosa', cmap='rgb')
plt.scatter(train_data[train_target['花の種類']==2]['がく片の長さ'], train_data[(train_target == 2).values]['花びらの幅'], label='setosa', cmap='rgb')

# グラフ幅
plt.xlim(3, 9)
plt.ylim(0, 5)

# ラベル
plt.xlabel('Length of sepal')
plt.ylabel('Width of sepal')

# グラフタイトル
plt.title('Relation between length and width of sepal')

# 出力
plt.legend()
```

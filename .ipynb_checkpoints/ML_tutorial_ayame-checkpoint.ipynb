```python
# library
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

# データセットを導入
from sklearn.datasets import load_iris                # アヤメデータ
from sklearn.model_selection import train_test_split  # 学習とテスト用のデータを分割するやつ
from sklearn.svm import LinearSVC                     # 学習モデル
from sklearn.metrics import accuracy_score            # 答え合わせ

flowers = load_iris()
```

```python
# pandasで整理
flowers_data = pd.DataFrame(flowers.data, columns=['がく片の長さ', 'がく片の幅', '花びらの長さ', '花びらの幅'])
flowers_target = pd.DataFrame(flowers.target, columns=['花の種類'])
flower_all = pd.concat([flowers_data, flowers_target], axis=1) # axis: 0-rows, 1-columns

# 出力(10行)
flower_all.head(10)
```

```python
# 出力(平均、最大値など)
flower_all.describe()
```

```python
# 学習用とテスト用にデータを分割
split_size = 0.33  #全体の33%をテスト用に使用
train_data, test_data, train_target, test_target = train_test_split(flowers_data, flowers_target, test_size=split_size)
```

```python
# 出力
train_data.head(5)
```

```python
train_target.head(5)
```

```python
train_data.describe()
```

```python
# train_target['花の種類']==0 と　train_data[(train_target == 0).values
# どちらを使用しても同じ
# 条件式が成り立つかがboolで返ってくる

# matplotlibで出力
plt.scatter(train_data[train_target['花の種類']==0]['がく片の長さ'], train_data[(train_target == 0).values]['がく片の幅'], label='setosa', cmap='rgb')
plt.scatter(train_data[train_target['花の種類']==1]['がく片の長さ'], train_data[(train_target == 1).values]['がく片の幅'], label='setosa', cmap='rgb')
plt.scatter(train_data[train_target['花の種類']==2]['がく片の長さ'], train_data[(train_target == 2).values]['がく片の幅'], label='setosa', cmap='rgb')

# グラフ幅
plt.xlim(3, 9)
plt.ylim(0, 5)

# ラベル
plt.xlabel('Length of sepal')
plt.ylabel('Width of sepal')

# グラフタイトル
plt.title('Relation between length and width of sepal')

# 出力
plt.legend()
```

```python
# matplotlibで出力
plt.scatter(train_data[train_target['花の種類']==0]['花びらの長さ'], train_data[(train_target == 0).values]['花びらの幅'], label='setosa', cmap='rgb')
plt.scatter(train_data[train_target['花の種類']==1]['花びらの長さ'], train_data[(train_target == 1).values]['花びらの幅'], label='setosa', cmap='rgb')
plt.scatter(train_data[train_target['花の種類']==2]['花びらの長さ'], train_data[(train_target == 2).values]['花びらの幅'], label='setosa', cmap='rgb')

# グラフ幅
plt.xlim(0, 9)
plt.ylim(0, 5)

# ラベル
plt.xlabel('Length of sepal')
plt.ylabel('Width of sepal')

# グラフタイトル
plt.title('Relation between length and width of sepal')

# 出力
plt.legend()
```

```python
# DataFrameの作成
gakuhen_train = train_data.loc[:,['がく片の長さ','がく片の幅']]
hanabira_train = train_data.loc[:,['花びらの長さ','花びらの幅']]

# 学習モデルを構築
gakuhen_model = LinearSVC()
hanabira_model = LinearSVC()

# モデルに学習させる
gakuhen_model.fit(gakuhen_train, train_target)
hanabira_model.fit(hanabira_train, train_target)

# 予測値を代入させる
gakuhen_predict = gakuhen_model.predict(test_data.loc[:,['がく片の長さ','がく片の幅']])
hanabira_predict = hanabira_model.predict(test_data.loc[:,['花びらの長さ','花びらの幅']])

## gakuhen_scoreとhanabira_scoreにそれぞれに結果を代入
gakuhen_score = accuracy_score(test_target, gakuhen_predict)
hanabira_score = accuracy_score(test_target, hanabira_predict)

# 出力(答え合わせ)
print('がく片の長さと幅コンビの正解率：{}'.format(gakuhen_score))
print('花びらの長さと幅コンビの正解率：{}'.format(hanabira_score))
```
